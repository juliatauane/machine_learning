**Importando Bibliotecas**

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import AdaBoostClassifier
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
class_le = LabelEncoder()
```

```python
#IRIS DATA SET
file_path = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
iris = pd.read_csv(file_path, header=None)
y_iris = class_le.fit_transform(iris[4].values)

X_iris = iris.iloc[:, range(0, 4)].values
y_iris
iris.rename(columns={0: "sepal length (cm)", 1: "sepal width (cm)", 2: "petal length (cm)", 3: "petal width (cm)", 4: "target"}, inplace=True)
iris.head()
```

```python
#WINE DATA SET
file_path = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
wine = pd.read_csv(file_path, header=None)
y_wine = class_le.fit_transform(wine[13].values)

X_wine = wine.iloc[:, range(0, 13)].values
y_wine
wine.rename(columns={0: "ID", 1: "Alcohol", 2: "Malic acid", 3: "Ash", 4: "Alcalinity of ash", 5: "Magnesium",
                 6: "Total phenols",7: "Flavanoids", 8: "Nonflavanoid phenols", 9: "Proanthocyanins", 10: "Color intensity", 11: "Hue", 12: "OD280/OD315 of diluted wines", 13: "Proline"}, inplace=True)
wine.head()
```

# **3. Gradient boosting**

[https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.GradientBoostingClassifier.html?highlight=gradient%20boosting#sklearn.ensemble.GradientBoostingClassifier](https://)

[https://towardsdatascience.com/machine-learning-part-17-boosting-algorithms-adaboost-in-python-d00faac6c464](https://)

[https://machinelearningmastery.com/gradient-boosting-with-scikit-learn-xgboost-lightgbm-and-catboost/](https://)

```python
# gradient boosting for classification in scikit-learn
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from matplotlib import pyplot

# evaluate the model
modelboosting = GradientBoostingClassifier()
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

# make a single prediction
row = [[2.56999479, -0.13019997, 3.16075093, -4.35936352, -1.61271951, -1.39352057, -2.48924933, -1.93094078, 3.26130366, 2.05692145]]
```

```python
#Iris data set
row_iris = [[2.56999479, -0.13019997, 3.16075093, -4.35936352]]
n_scores_iris = cross_val_score(modelboosting, X_iris, y_iris, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
print('Accuracy Iris Gradient Boosting: %.4f (%.4f)' % (mean(n_scores_iris), std(n_scores_iris)))

# fit the model on the whole dataset
modelboosting.fit(X_iris, y_iris)

# make a single prediction
yhat_iris = modelboosting.predict(row_iris)
print('Prediction: %d' % yhat_iris[0])
```

```python
#Wine data set
# make a single prediction
row_wine = [[2.56999479, -0.13019997, 3.16075093, -4.35936352, -1.61271951, -1.39352057, -2.48924933, -1.93094078, 3.26130366, 2.05692145, 2.56999479, -0.13019997, 3.16075093]]
n_scores_wine = cross_val_score(modelboosting, X_wine, y_wine, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
print('Accuracy Wine Gradient Boosting: %.4f (%.4f)' % (mean(n_scores_wine), std(n_scores_wine)))
# fit the model on the whole dataset
modelboosting.fit(X_wine, y_wine)

# make a single prediction
yhat_wine = modelboosting.predict(row_wine)
print('Prediction: %d' % yhat_wine[0])
```

```python
#Glass data set
row_glass = [[2.56999479, -0.13019997, 3.16075093, -4.35936352, -1.61271951, -1.39352057, -2.48924933, -1.93094078, 3.26130366]]
n_scores_glass = cross_val_score(modelboosting, X_glass, y_glass, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
print('Accuracy Glass Gradient Boosting: %.4f (%.4f)' % (mean(n_scores_glass), std(n_scores_glass)))

# fit the model on the whole dataset
modelboosting.fit(X_glass, y_glass)

# make a single prediction
yhat_glass = modelboosting.predict(row_glass)
print('Prediction: %d' % yhat_glass[0])
```
